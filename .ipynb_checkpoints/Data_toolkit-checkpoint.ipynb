```python

# Explain Np.array , np.asarray, np.asanyarray 
# np.array 
#  Purpose: To create a new array from any object exposing the array interface, a nested sequence, or an array.
# Behavior: Always returns a new array.
# Parameters:
# object: The data to be converted to an array.
# dtype: Desired data type of the array. If not given, the type will be inferred from the data.
import numpy as np
data = [1,2,3]
arr = np.array(data)
print(arr)

# np.asaarray 
# Converts input to an array only if it isn't already one.
# Does not copy data if the input is already an of correct type 
data = [ 1,2,3]
array = np.asarray(data)
print(array)

# np.asanyarray
# Converts input to an array or passes thorugh array subclasses.
```

```python
# Difference betweeen deep copy and shallow copy
#  A shallow copy creates a new object, but inserts references into it to the objects found in the original. It does not create copies of the nested objects
import copy
original = [[1,2,3],[4,5,6]]
shallow_copy = copy.copy(original)
shallow_copy[0][1] = 20
print("original:", original)
print("Shallow copy:", shallow_copy)
```

```python
# Deep copy: Deep copy creates a new object and recursively copies all the objecs contained withing the orginal , creating independent copies of these nested objects.

import copy
original = [[1,2,3],[4,5,6]]
deep_copy = copy.deepcopy(original)
deep_copy[0][1] = 20
print("original:", original)
print("deep copy:", deep_copy)
```

```python
# Clean and transform the "phone" column in the sample dataset to remove non-numeric characters and convert it into the  numeric data type. Display the table attribute and data type of each column
import pandas as pd 

# Read the CSV file into a DataFrame
df = pd.read_csv("People Data.csv")

# Print the original DataFrame
print("Original DataFrame:")
print(df)

# Clean the "Phone" column
df['Phone'] = df['Phone'].str.replace(r'\D','', regex=True)

# Convert the "Phone" column to numeric
df['Phone'] = pd.to_numeric(df['Phone'], errors='coerce')

# Print the altered DataFrame
print('\nAltered DataFrame:')
print(df)

# Print the data types of each column
print("\nData Types:")
print(df.dtypes)
```

```python
# Create a time-series plot dataset in a pandas Data frame with columns Data , Temperature , HUmidity and perform following operations 
# A. Plot the the tmeperature abd humidity on the same plot with diffrent y-axes (left y-axis for temparature and right y-axes for HUmidity.
import pandas as pd
import numpy as np

# Create a date range
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Generate random data for temperature and humidity
np.random.seed(0)
temperature = np.random.uniform(low=20, high=35, size=100)
humidity = np.random.uniform(low=40, high=90, size=100)

# Create a DataFrame
data = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Set the 'Date' column as the index
data.set_index('Date', inplace=True)

# Print the first few rows of the DataFrame
print(data.head())
```

```python
import matplotlib.pyplot as plt

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot temperature data
ax1.plot(data.index, data['Temperature'], 'r-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Create a second y-axis for humidity
ax2 = ax1.twinx()
ax2.plot(data.index, data['Humidity'], 'b-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# SEt the title Temperature and Humidity Over Time
plt.title('Temperature and Humidity Over Time')

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
# Create a numpy array data containing 1000 sample from normal distribution. Perform the following takd using matplotlib.
# a. PLot a histogram of 30 bin 
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.normal(0, 1, 1000)

# Plot the histogram
plt.hist(data, bins=30, edgecolor='black', density=True, alpha=0.5)

# Calculate the mean and standard deviation of the data
mean = np.mean(data)
std = np.std(data)

# Generate x-values for the normal distribution line
x = np.linspace(min(data), max(data), 1000)

# Calculate the y-values for the normal distribution line
y = (1 / (np.sqrt(2 * np.pi * std**2))) * np.exp(-(x - mean)**2 / (2 * std**2))

# Plot the normal distribution line
plt.plot(x, y, 'r-', linewidth=2, label='Normal Distribution')

plt.title('Histogram with PDF overlay')
plt.xlabel('Value')
plt.ylabel('Probability / Density')
plt.legend()
plt.show()
```

```python
# Create a seaborn scatter plot of two random arrays , color points based on their position relative to the origin (quadrants), add a legend, lable the axes and set the title as Quadrant-wise Scatter Plot'
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Generate two random arrays
x = np.random.randint(-10, 11, 50)
y = np.random.randint(-10, 11, 50)

# Create a scatter plot
plt.figure(figsize=(8, 6))
data = {'x': x, 'y': y}  # Create a dictionary with x and y data
ax = sns.scatterplot(data=data, x='x', y='y', s=100)

# Color the points based on their quadrants
quadrants = []
for i, j in zip(x, y):
    if i > 0 and j > 0:
        quadrants.append('Q1')
    elif i < 0 and j > 0:
        quadrants.append('Q2')
    elif i < 0 and j < 0:
        quadrants.append('Q3')
    else:
        quadrants.append('Q4')

palette = {'Q1': 'r', 'Q2': 'g', 'Q3': 'b', 'Q4': 'y'}
ax.set_palette(palette)
for i, quad in enumerate(quadrants):
    ax.scatter(x[i], y[i], color=palette[quad])

# Add legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles, labels=['Q1', 'Q2', 'Q3', 'Q4'], title='Quadrants', loc='upper left')

# Label the axes and set the title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

```python
# With Bokeh plot a line chart of a sine wave function, add grid lines, label the axis and set the title as Sine Wave function.
from bokeh.plotting import figure, show
import numpy as np
from bokeh.io import show, output_file
# Generate data for the sine function
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Function", x_axis_label="x", y_axis_label="sin(x)")

# Add the sine function to the plot
p.line(x, y, line_width=2)
output_file("sine.html")
# Show the plot
show(p)
```
